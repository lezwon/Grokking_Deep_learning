```python
weights = [ 0.1, 0.4, 0.6 ]
inputs = [0.4, 0.6, 0.1]
goal_predictions = [[1,0,0],[0,1,1],[0,0,1]]
alpha = 0.001
```

```python
def neural_network(input, weights):
    return element_multiplication(input, weights)

def element_multiplication(input, weights):
    output = []
    for weight in weights:
        output.append(input * weight)
    return output
```

```python
errors = [2, 2, 2]
deltas = [0, 0, 0]
while sum(errors) > 0.001:
    preds = neural_network(inputs[0], weights)
    for idx, pred in enumerate(preds):
        deltas[idx] = (pred - goal_predictions[0][idx])
        errors[idx] = deltas[idx] ** 2

    weight_deltas = element_multiplication(inputs[0], deltas)

    for idx, weight_delta in enumerate(weight_deltas):
        weights[idx] -= weight_delta * alpha

    print (sum(errors))
```
