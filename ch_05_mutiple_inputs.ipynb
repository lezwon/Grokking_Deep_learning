```python
inputs0 = [3, 6, 5, 2]
inputs1 = [0.5, 0.6, 0.1, 0.9]
inputs2 = [0.3, 0.1, 0.2, 0.8]

weights = [0.3, 0.6, 0.1]

goal = [1, 0, 1, 0]

alpha = 0.0001
```

```python
def neural_network(weights, input):
    return weight_sum(weights, input)

def weight_sum(weights, input):
    assert(len(weights) == len(input))
    output = 0
    for index in range(len(input)):
        output += weights[index] * input[index]
    return output

def get_weight_deltas(delta, input):
    weight_deltas = []
    for value in input:
        weight_deltas.append(value * delta)
    return weight_deltas
```

```python
# variable = for [input1, input2, input3] in zip(inputs0, inputs1, inputs2)
inputs = [list(a) for a in zip(inputs0, inputs1, inputs2)]


```

```python
error = 2
while error > 0.000001:
    for input in inputs:
        pred = neural_network(weights, input)
        delta = pred - goal[0]
        error = delta ** 2
        weight_deltas = get_weight_deltas(delta, input)
#         weight_deltas[1] = 0
        for idx, weight_delta in enumerate(weight_deltas):
        #     print (weight, weight_delta)
            weights[idx] -= weight_delta * alpha

        print(pred)

    print(weights)
```
